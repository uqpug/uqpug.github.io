```python
import requests
from bs4 import BeautifulSoup
```

```python
# Set up soup
url = "https://en.wikipedia.org/wiki/University_of_Queensland"

website = requests.get(url)

soup = BeautifulSoup(website.content)
```

```python
# Examine HTML

soup.prettify()
```

```python
days = soup.find_all("div", class_ = "current-events-content description")

for day in days:
    for topic in day.find_all("ul"):
        for headline in topic.children:
            print(headline)
```
