### Python in Windows PowerShell
How do I run python through Windows PowerShell?

#### Answers
Assuming you have Python installed *in a location that your computer knows about* (not always an obvious task), then run

```bash
python
```

in a PowerShell prompt to launch the interpreter.

A few things to note ([see the docs](https://docs.python.org/3/using/cmdline.html) for details)
* Just `python` will launch the interpreter, where you run code line-by-line
* If you have a Python script called "scriptname.py" (or any other name), `python scriptname.py` will run the file
* `python -c ...` runs Python commands manually sent in 
* `python -m ...` runs a module. For example, `python -m pip install ...` runs the `pip` module (useful if your computer can't find `pip` but *can* find `python`)
